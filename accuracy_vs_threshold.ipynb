```python
threshold = [20, 30, 40, 50, 60, 70, 80, 90, 99]
screen1 = [99.7, 99.7, 99.41, 99.1, 98.8, 97, 94, 78, 48]
screen2 = [100, 100, 100, 99, 98, 98, 97, 85, 46 ]
screen3 = [100, 100, 100, 99, 98, 98,  97.5, 88, 85]

fig, axs = plt.subplots(nrows = 3, ncols = 1, figsize =(10,10))

color = ['r', 'c', 'g'] 
for i, ax in enumerate(fig.axes, start=1):
    df = globals()[f'screen{i}']                    
    ax.plot(threshold,df, c=color[i-1])   
    ax.scatter(threshold,df, s=10, c=color[i-1], marker='*')
    ax.set_xlabel('Threshold')
    ax.set_ylabel('Accuracy')
    ax.grid(True)
    ax.set_title(f'Subfigure for accuracy detection of computer screen{i}', fontsize = 15)
fig.tight_layout()
```
