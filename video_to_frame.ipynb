```python
import cv2
import os

vid = cv2.VideoCapture('obj_det_100.avi')
currentFrame = 0

if not os.path.exists('drive100'):
    os.makedirs('drive100')
    
    

while(True):
    successs, frame = vid.read()
    
    #cv2.imshow("Output", frame)
    cv2.imwrite('./drive100/frame' + str(currentFrame) + '.jpg', frame)
    currentFrame +=1
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
vid.release()
cv2.closeALLWindows()
```
